# 2. 조립설명서 부품 - CAD 매칭 정확도

### 최소 10개 이상 부품에 대하여, 부품 - CAD 매칭을 성공할 확률을 10번 수행하여 평균 정확도를 측정한다.

#### 부품: 11가지, Test 이미지: 110장 (부품 별 10장), CAD 모델: 11개 (부품 별 1개)

### 1) Dataset (query 부품 이미지와 CAD 파일을 projection한 이미지)

#### 전체 부품의 종류
###### 이름: (가구이름)_step# 혹은 (가구이름)_part#      step: 중간산출물, part: 기본부품

```python
import matplotlib.pyplot as plt
import glob, os
%matplotlib inline

data_dir = '../data'
img_dir = os.path.join(data_dir, 'IMAGES')
view_dir = os.path.join(data_dir, 'VIEWS')

img_class_dirs = glob.glob(os.path.join(img_dir, '*'))
view_class_dirs = glob.glob(os.path.join(view_dir, '*'))

classes = ''
for dir in img_class_dirs:
    classes += os.path.basename(dir)+', '

classes = classes[0:-1]
print('Total %d classes :: %s' % (len(img_class_dirs), classes))
```

#### 부품의 예시. 윗줄: 부품 이미지, 아랫줄: CAD 모델 projection 이미지

```python
classes = img_class_dirs[0:10:2]
imgs = []
views = []
for cl in classes:
    filenames = glob.glob(os.path.join(cl, 'test/*.png'))
    imgs.append(plt.imread(filenames[0]))
    filenames = glob.glob(os.path.join(view_dir, os.path.basename(cl), '*.png'))
    views.append(plt.imread(filenames[0]))

fig, ax = plt.subplots(2, 5, sharex=True, sharey=True, figsize=(100,40))
for i in range(5):
    ax[0][i].imshow(imgs[i])
    ax[1][i].imshow(views[i])
    ax[0][i].set_title(os.path.basename(classes[i]), fontsize=100, color='#03719c')
    if i==0:
        ax[0][i].text(0., 90., "IMAGE", rotation=90, fontsize=90, verticalalignment='center', color='#009337')
        ax[1][i].text(0., 100., "CAD", rotation=90, fontsize=90, verticalalignment='center', color='#009337')


plt.xticks([])
plt.yticks([])
plt.subplots_adjust(hspace=0.001, wspace=0.0001)
```
