#### 최소 10개 이상 부품에 대하여, 부품 - CAD 매칭을 성공할 확률을 10번 수행하여 평균 정확도를 측정한다.

#### 입력 부품 이미지와 CAD 파일을 projection한 이미지 예시

```python
# input image - View pairs
import matplotlib.pyplot as plt
import glob, os

data_dir = '../data'
img_dir = os.path.join(data_dir, 'IMAGES')
view_dir = os.path.join(data_dir, 'VIEWS')

img_class_dirs = glob.glob(os.path.join(img_dir, '*'))
view_class_dirs = glob.glob(os.path.join(view_dir, '*'))

classes = ''
for dir in img_class_dirs:
    classes += os.path.basename(dir)+', '

classes = classes[0:-1]
print('Total %d classes :: %s' % (len(img_class_dirs), classes))
```

```python
classes = img_class_dirs[0:10:2]
imgs = []
views = []
for cl in classes:
    filenames = glob.glob(os.path.join(cl, 'test/*.png'))
    imgs.append(plt.imread(filenames[0]))
    filenames = glob.glob(os.path.join(view_dir, os.path.basename(cl), '*.png'))
    views.append(plt.imread(filenames[0]))

fig, ax = plt.subplots(2, 5, sharex=True, sharey=True, figsize=(100,40))
# ax[0].set_title("Image")
# ax[0].tick_params(labelcolor=(1.,1.,1.,0.0), top='off', bottom='off', left='off', right='off')
# ax[1].set_title("CAD")
# ax[1].tick_params(labelcolor=(1.,1.,1.,0.0), top='off', bottom='off', left='off', right='off')


# for i in range(1,6):
#     ax0 = fig.add_subplot(1,i,2*(i-1)+1)
#     ax1 = fig.add_subplot(2,i,2*i)
#     ax0.imshow(imgs[i])
#     ax1.imshow(views[i])
#     ax0.set_title(os.path.basename(classes[i]), fontsize=100)
    
for i in range(5):
    ax[0][i].imshow(imgs[i])
    ax[1][i].imshow(views[i])
    ax[0][i].set_title(os.path.basename(classes[i]), fontsize=100)


temp=plt.xticks([])
temp=plt.yticks([])
plt.subplots_adjust(hspace=0.0001, wspace=0.0001)
```
