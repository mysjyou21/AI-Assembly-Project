### 시험에 사용된 조립설명서 10장

```python
import glob, os
input_img_dir = './input/test/cuts'
input_img_paths = sorted(glob.glob(os.path.join(input_img_dir, '*.png')))
input_img_names = [os.path.basename(path) for path in input_img_paths]

print('Total %d assembly images:' % len(input_img_names))
print(*input_img_names, sep=', ')
```

### 객체의 종류
* 딥러닝 기반 인식 대상: 6가지
  + 연결자 (Elements)
  + FP 유발자 (FPTrigger)
  + 원형 말풍선 (Guidance_Circle)
  + 사각형 말풍선 (Guidance_Square)
  + 나무 부품 (Parts)
  + 도구 (Tool)  
<br>
* Connected component 기반 인식 대상: 1가지
  + 부품번호 및 수량을 나타내는 영역  
<br>
* OCR 기반 인식 대상: 1가지
  + 부품번호 및 수량


### 성공률 산출 방법
딥러닝 기반 인식 대상 객체들의 mAP, 부품번호 및 수량 영역의 재현율(=Recall), 부품번호 및 수량의 OCR 인식률의 가중 평균을 성공률으로 한다.  
<br>
\begin{equation*}
score=\frac{1}{8} \Bigl(\sum_{i=1}^6 (AP)_i +\ Recall +\ OCR\;recognition\;rate\Bigr)
\end{equation*}
